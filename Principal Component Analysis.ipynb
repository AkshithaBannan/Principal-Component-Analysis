```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA()
pca.fit(scaled_data)
```

```python
# Scree plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

# Cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Convert to DataFrame
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
print(pca_df.head())
```

```python
from sklearn.cluster import KMeans

# Apply K-means clustering
kmeans = KMeans(n_clusters=3)  # Example with 3 clusters
kmeans.fit(scaled_data)
clusters = kmeans.labels_

# Add cluster labels to the original dataframe
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering (Original Data)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Silhouette score
sil_score = silhouette_score(scaled_data, clusters)
print('Silhouette Score:', sil_score)

# Davies–Bouldin index
db_index = davies_bouldin_score(scaled_data, clusters)
print('Davies–Bouldin Index:', db_index)
```

```python
# Apply K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3)
kmeans_pca.fit(pca_data)
pca_clusters = kmeans_pca.labels_

# Add cluster labels to the PCA dataframe
pca_df['Cluster'] = pca_clusters
```

```python
# Apply K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3)
kmeans_pca.fit(pca_data)
pca_clusters = kmeans_pca.labels_

# Add cluster labels to the PCA dataframe
pca_df['Cluster'] = pca_clusters
```

```python
# Silhouette score for PCA data
sil_score_pca = silhouette_score(pca_data, pca_clusters)
print('Silhouette Score (PCA Data):', sil_score_pca)

# Davies–Bouldin index for PCA data
db_index_pca = davies_bouldin_score(pca_data, pca_clusters)
print('Davies–Bouldin Index (PCA Data):', db_index_pca)
```

```python
# Print and compare Silhouette scores and Davies–Bouldin indices
print('Original Data: Silhouette Score =', sil_score, ', Davies–Bouldin Index =', db_index)
print('PCA Data: Silhouette Score =', sil_score_pca, ', Davies–Bouldin Index =', db_index_pca)
```

```python
# Summarize observations 
print("The Silhouette Score for the PCA data is higher than the original data, indicating better clusters.")
print("The Davies–Bouldin Index for the PCA data is lower than the original data, indicating less compact clusters.")
```

```python
# Discuss the impact
print("Dimensionality reduction with PCA helped in reducing noise and improving cluster definition, as evidenced by the Silhouette Score and Davies–Bouldin Index.")
print("However, some information might be lost during the transformation, which could impact clustering results differently.")
```
